```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/1. Weather Data.csv", parse_dates=['Date'], index_col='Date')
print(df.head())

# Plotting Original Data for the Entire Dataset
plt.figure(figsize=(12, 6))
plt.plot(df['Temperature'], label='Original Temperature', color='blue')
plt.title('Original Weather Temperature (Entire Dataset)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()

# Applying Moving Average
window_size = 7
df['Moving_Avg'] = df['Temperature'].rolling(window=window_size).mean()

# Plotting Original Data and Moving Average
plt.figure(figsize=(12, 6))
plt.plot(df['Temperature'], label='Original Temperature', color='blue', alpha=0.6)
plt.plot(df['Moving_Avg'], label=f'{window_size}-day Moving Average', color='red', linewidth=2)
plt.title(f'Original Temperature and {window_size}-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```
