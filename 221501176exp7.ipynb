```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv(r"/content/sample_data/1. Weather Data.csv", parse_dates=['Date'], index_col='Date')
print(df.head())

result = seasonal_decompose(df['Temperature'], model='additive', period=365)

plt.figure(figsize=(12, 8))

# Plot the original time series
plt.subplot(4, 1, 1)
plt.plot(df['Temperature'], label='Original Data')
plt.title('Original Temperature Time Series')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')

# Plot the trend component
plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.xlabel('Date')
plt.ylabel('Trend')

# Plot the seasonal component
plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.title('Seasonal Component')
plt.xlabel('Date')
plt.ylabel('Seasonality')

# Plot the residual component
plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residuals', color='red')
plt.title('Residual Component')
plt.xlabel('Date')
plt.ylabel('Residual')

plt.tight_layout()
plt.show()
```
